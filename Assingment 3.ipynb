# Part One

```python
#Explore the data#Read iris.txt into a data frame (2 point)

import pandas as pd
from pandas.tools.plotting import scatter_matrix
%matplotlib inline
DT = pd.read_csv("iris.txt")
```


# Part Two

```python
#Explore the data#Read iris.txt into a data frame (2 point)

#Print the head the dataset (2 point)
print (DT.head())
#Print the tail the dataset (2 point)
print (DT.tail())
#Print summary statistics for the data frame (both numeric and categorical columns) (2 point)
print(DT.describe())
#Generate a box plot for the data set (2 point)
DT.plot.box()
#Generate a line plot the dataset (2 point)
DT.plot.line()
#Generate a histogram of the dataset (2 point) 
DT.plot.hist()
```


# Interpret Results

    According to the results there is not much variance in the Sepal Width of the flowers, but we do have some variation in the pedal's length. A good amount of the Sepal length seems to be under 3.5 

# Part Three

```python
#Group the data on the Species column and explore the partitioned dataset

#Print the summary statistics for each species in the dataset (2 point)
Darwin= DT.groupby("Species")
print(Darwin.describe())

#Generate a boxplot for each species in the dataset (2 point)
print(Darwin.plot.box())

#Generate a scatter matrix for the dataset (2 point)
scatter_matrix(DT, alpha=0.4, figsize=(7, 7), diagonal='hist')
```
